# 3.  Preprocesamiento de datos

```python
# Importaci√≥n de librerias
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
```

```python
# Lectura del dataset
telecom_data = pd.read_csv(
    "c:/Users/cary_/OneDrive/Escritorio/Data analyst/Sprint 14/telecom_clients_us.csv", sep=",")

clients_data = pd.read_csv(
    "c:/Users/cary_/OneDrive/Escritorio/Data analyst/Sprint 14/telecom_dataset_us.csv", sep=",")
```

```python
# Estudia tipo de datos del dataset
clients_data.info()
print("\n")
telecom_data.info()
```

Con la informaci√≥n observada en los datasets, se puede notar que algunas columnas no tienen los tipos de datos correctos. Lo primero ser√° revisar los valores duplicados y ausentes para poder hacer las conversiones a los tipos de datos necesarios. A continuaci√≥n, se usar√° una funci√≥n que me ayudar√° a convertir algunos tipos de datos y reducir el tama√±o de la memoria usada por ambos datasets.

```python
# Revisar si existe alg√∫n valor duplicado
print(f'Duplicados totales para clients_data: {clients_data.duplicated().sum()}')
print(f'Duplicados totales para telecom_data: {telecom_data.duplicated().sum()}')
print("\n")

# Borrar valores duplicados
clients_data = clients_data.drop_duplicates().reset_index(drop=True)
```

En esta secci√≥n, los resultados muestran que existen valores ausentes en las columnas internal y operator_id. Dado que la columna internal solo contiene valores como True y False, utilizar√© False para rellenar los valores ausentes. Para la columna operator_id, debido a que es informaci√≥n sensible y no tengo forma de consultar al ingeniero de datos sobre la informaci√≥n faltante, rellenar√© los valores ausentes con 000000.

```python
# Revisar valores ausentes en cada dataset
print("Valores nulos totales para clients_data: \n", clients_data.isna().sum())
print("\n")
print("Valores nulos totales para telecom_data: \n", telecom_data.isna().sum())

# Se usa el metodo fillna para rellenar valores ausentes
clients_data['internal'] = clients_data['internal'].fillna(False).infer_objects(copy=False)
clients_data['operator_id'] = clients_data['operator_id'].fillna(000000)
```

```python
# Impresi√≥n para comprobar que los valores nulos ya no estan
print("\n")
print("\nValores nulos totales para clients_data: \n", clients_data.isna().sum())
```

```python
# Impresi√≥n de muestra de datos del dataset telecom_clients_us
display(clients_data.head())

print("\n")

# Impresi√≥n de muestra de datos del dataset telecom_dataset_us
display(telecom_data.head())
```

```python
# Funci√≥n para convertir los tipos datos de ambos datasets y bajar el tama√±o de la memoria usada
def downcast_dtypes(df):
    float_cols = df.select_dtypes("float").columns
    int_cols = df.select_dtypes("integer").columns

    df[float_cols] = df[float_cols].apply(pd.to_numeric, downcast="float")
    df[int_cols] = df[int_cols].apply(pd.to_numeric, downcast="integer")

    return df

# Aplicaci√≥n de la funcion a ambos datasets
print(downcast_dtypes(clients_data))
print("\n")
print(downcast_dtypes(telecom_data))
```

Despu√©s de aplicar la funci√≥n, se puede observar que a√∫n existen algunas columnas que es necesario convertir. En cuanto a los nombres de las columnas, no ser√° necesario hacer cambios ya que todas est√°n en min√∫sculas, tienen nombres descriptivos y siguen el formato snake case.

```python
# Impresion de los datasets para revisar nuevamente los tipos de datos 
clients_data.info()
print("\n")
telecom_data.info()
```

```python
# Conversi√≥n de tipos de datos

# Se convierte a tipo fecha la columna date_start del dataset telecom_data
telecom_data['date_start'] = pd.to_datetime(telecom_data['date_start'])

# Se convierte a tipo category columna tariff_plan del dataset telecom_data
telecom_data['tariff_plan'] = telecom_data['tariff_plan'].astype('category')

# Se convierte a tipo fecha la columna date_start del dataset clients_data
clients_data['date'] = pd.to_datetime(clients_data['date'])

# Se convierte a tipo bool columna internal del dataset clients_data
clients_data['internal'] = clients_data['internal'].astype('bool')

# Se convierte a tipo int64 columna operator_id del dataset clients_data
clients_data['operator_id'] = clients_data['operator_id'].astype('int64')
```

### Analizar la relaci√≥n entre n√∫mero de llamadas, duraci√≥n de llamadas, llamadas perdidas para cada operador.

```python
# Eliminar filas con operator_id igual a 0 
clients_data_clean = clients_data.loc[clients_data['operator_id'] != 0]

# Agrupar los datos por operador y calcular las m√©tricas
metrics = clients_data_clean.groupby('operator_id').agg(
    total_calls=('calls_count', 'sum'),
    total_call_duration=('call_duration', 'sum'),
    missed_calls=('is_missed_call', 'sum')
).reset_index()


# Crear gr√°ficos de viol√≠n para cada m√©trica
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Gr√°fico para el total de llamada
sns.violinplot(x='total_calls', data=metrics, ax=axs[0])
axs[0].set_title('Distribuci√≥n del N√∫mero Total de Llamadas por Operador')

# Gr√°fico para la duraci√≥n total de llamadas
sns.violinplot(x='total_call_duration', data=metrics, ax=axs[1])
axs[1].set_title('Distribuci√≥n de la Duraci√≥n Total de las Llamadas por Operador')

# Gr√°fico para el total de llamadas perdidas
sns.violinplot(x='missed_calls', data=metrics, ax=axs[2])
axs[2].set_title('Distribuci√≥n del N√∫mero de Llamadas Perdidas por Operador')

plt.tight_layout()
plt.show()
```

Distribuci√≥n del N√∫mero Total de Llamadas por Operador

El gr√°fico de viol√≠n muestra que la mayor√≠a de los operadores tienen un n√∫mero total de llamadas relativamente bajo, con una concentraci√≥n alrededor de los valores m√°s peque√±os. Sin embargo, hay una cola larga hacia valores m√°s altos, lo que indica que hay algunos operadores que manejan un n√∫mero significativamente mayor de llamadas.

Distribuci√≥n de la Duraci√≥n Total de las Llamadas por Operador

La distribuci√≥n de la duraci√≥n total de las llamadas es similar a la del n√∫mero total de llamadas. La mayor√≠a de los operadores tienen una duraci√≥n total de llamadas baja, pero hay una cola larga hacia duraciones m√°s largas. Esto sugiere que, aunque la mayor√≠a de los operadores no manejan llamadas muy largas, hay algunos que s√≠ tienen duraciones significativamente mayores.

Distribuci√≥n del N√∫mero de Llamadas Perdidas por Operador

El n√∫mero de llamadas perdidas tambi√©n muestra una concentraci√≥n alrededor de valores bajos, con la mayor√≠a de los operadores teniendo pocas llamadas perdidas. Sin embargo, tambi√©n hay una cola larga, indicando que algunos operadores tienen un n√∫mero relativamente alto de llamadas perdidas.

### Identificar la relacion entre el numero de llamadas y la duracion de la llamada.

```python
# Agrupar los datos por operador y calcular las m√©tricas
metrics = clients_data_clean.groupby('operator_id').agg(
    total_calls=('calls_count', 'sum'),
    total_call_duration=('call_duration', 'sum')
).reset_index()

# Calcular la correlaci√≥n
correlation, p_value = pearsonr(metrics['total_calls'], metrics['total_call_duration'])
print(f'Correlaci√≥n: {correlation}, Valor p: {p_value}')

# Crear un gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_calls', y='total_call_duration', data=metrics)
plt.title('Relaci√≥n entre N√∫mero Total de Llamadas y Duraci√≥n Total de Llamadas')
plt.xlabel('N√∫mero Total de Llamadas')
plt.ylabel('Duraci√≥n Total de Llamadas (minutos)')

# Ajustar una l√≠nea de regresi√≥n
sns.regplot(x='total_calls', y='total_call_duration', data=metrics, scatter=False, color='red')
plt.show()
```

Pendiente Positiva: La l√≠nea de regresi√≥n tiene una pendiente positiva, lo que indica que hay una relaci√≥n positiva entre el n√∫mero total de llamadas y la duraci√≥n total de las llamadas. Esto significa que, en general, a medida que aumenta el n√∫mero de llamadas realizadas por un operador, tambi√©n aumenta la duraci√≥n total de estas llamadas.

Intervalo de Confianza: El √°rea sombreada alrededor de la l√≠nea roja representa el intervalo de confianza. Nos da una idea de la fiabilidad de la l√≠nea de regresi√≥n. En este caso, el intervalo es relativamente estrecho, lo que sugiere que la relaci√≥n observada es bastante consistente.

Fuerza de la Relaci√≥n: Un valor de correlaci√≥n de 0.9412309096318842 indica una relaci√≥n muy fuerte entre las dos variables. Esto significa que, en general, a medida que el n√∫mero total de llamadas aumenta, la duraci√≥n total de las llamadas tambi√©n aumenta significativamente.

### Comparar el tiempo de duracion de las llamadas para cada operador.

```python
# Agrupar los datos por operador y calcular la duraci√≥n promedio de las llamadas
metrics = clients_data_clean.groupby('operator_id').agg(
    avg_call_duration=('call_duration', 'mean')
).reset_index()

# Ordenar los datos por la columna 'avg_call_duration' en orden descendente 
metrics_sorted = metrics.sort_values(by='avg_call_duration', ascending=True).reset_index(drop=True)

# Mostrar los resultados
print(metrics_sorted.head(10))

# Filtrar operadores con duraci√≥n promedio de llamadas mayor a 0 
metrics_filtered = metrics[metrics['avg_call_duration'] > 0]

# Crear gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(metrics_filtered['operator_id'], metrics_filtered['avg_call_duration'], marker='o', linestyle='-', color='b')
plt.title('Duraci√≥n Promedio de las Llamadas por Operador')
plt.xlabel('ID del Operador')
plt.ylabel('Duraci√≥n Promedio de Llamadas (minutos)')
plt.grid(True)
plt.show()
```

La mayor√≠a de los operadores tienen una duraci√≥n promedio de llamadas relativamente baja, con muchos puntos concentrados por debajo de los 2,000 minutos.

Sin embargo, hay varios valores at√≠picos (outliers) donde la duraci√≥n promedio de las llamadas supera los 10,000 minutos. Estos operadores tienen duraciones de llamadas significativamente m√°s largas en comparaci√≥n con los dem√°s.

Los operadores con duraciones promedio de llamadas superiores a 10,000 minutos se destacan claramente en el gr√°fico. Estos valores at√≠picos pueden ser indicativos de varios factores, como una mayor carga de trabajo, llamadas m√°s complejas o posibles ineficiencias operativas.

Los operadores con duraciones promedio de llamadas muy altas pueden estar manejando consultas m√°s complicadas o tener una mayor carga de trabajo, lo que resulta en tiempos de llamada m√°s largos.Tambi√©n puede sugerir que algunos operadores son menos eficientes, lo que podr√≠a ser una √°rea para mejorar.

Es posible que ciertos operadores tengan patrones de uso distintos que los llevan a tener duraciones de llamadas significativamente m√°s largas.

### Determinar las m√©tricas clave para definir la ineficacia de los operadores (llamadas perdidas, tiempo de espera, n√∫mero de llamadas entrantes y salientes)

```python
# Calcular la cantidad de llamadas perdidas
llamadas_perdidas = clients_data_clean[clients_data_clean['is_missed_call'] == True].groupby('operator_id')['calls_count'].sum().reset_index()
llamadas_perdidas.rename(columns={'calls_count': 'llamadas_perdidas'}, inplace=True)

# Calcular el tiempo de espera promedio para llamadas entrantes
clients_data_clean['tiempo_espera'] = clients_data_clean['total_call_duration'] - clients_data_clean['call_duration']
tiempo_espera_promedio = clients_data_clean[clients_data_clean['direction'] == 'in'].groupby('operator_id')['tiempo_espera'].mean().reset_index()
tiempo_espera_promedio.rename(columns={'tiempo_espera': 'tiempo_espera_promedio'}, inplace=True)

# Calcular la cantidad de llamadas entrantes y salientes
llamadas_entrantes = clients_data_clean[clients_data_clean['direction'] == 'in'].groupby('operator_id')['calls_count'].sum().reset_index()
llamadas_entrantes.rename(columns={'calls_count': 'llamadas_entrantes'}, inplace=True)

llamadas_salientes = clients_data_clean[clients_data_clean['direction'] == 'out'].groupby('operator_id')['calls_count'].sum().reset_index()
llamadas_salientes.rename(columns={'calls_count': 'llamadas_salientes'}, inplace=True)

# Combinar todas las m√©tricas en un solo dataframe
metricas = llamadas_perdidas.merge(tiempo_espera_promedio, on='operator_id', how='outer')
metricas = metricas.merge(llamadas_entrantes, on='operator_id', how='outer')
metricas = metricas.merge(llamadas_salientes, on='operator_id', how='outer')

# Rellenar valores NaN con 0
metricas.fillna(0, inplace=True)

# Lista de columnas y sus respectivos t√≠tulos y etiquetas
columns = [
    ('llamadas_perdidas', 'Distribuci√≥n de llamadas perdidas por operador', 'Llamadas perdidas'),
    ('tiempo_espera_promedio', 'Distribuci√≥n del tiempo de espera de llamadas por operador', 'Tiempo de Espera Promedio'),
    ('llamadas_entrantes', 'Distribuci√≥n de llamadas entrantes por operador', 'Llamadas entrantes'),
    ('llamadas_salientes', 'Distribuci√≥n de llamadas salientes por operador', 'Llamadas salientes')
]

# Ciclo for para generar los gr√°ficos
for column, title, ylabel in columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='operator_id', y=column, data=metricas, alpha=0.5)
    plt.xlabel('Operador ID')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45)  # Rotar etiquetas del eje x si es necesario para mejor legibilidad
    plt.show()
```

Distribuci√≥n de llamadas perdidas por operador: La mayor√≠a de los operadores tienen un n√∫mero bajo de llamadas perdidas. Sin embargo, hay algunos operadores con un n√∫mero significativamente alto de llamadas perdidas, lo que indica problemas en su desempe√±o. Estos operadores pueden necesitar entrenamiento adicional o soporte para mejorar su eficiencia.

Distribuci√≥n del Tiempo de Espera de Llamadas por Operador: Un punto en el gr√°fico representa un operador espec√≠fico y su tiempo de espera promedio. Si un operador tiene un tiempo de espera promedio alto (puntos m√°s arriba en el eje Y), esto sugiere que las llamadas no est√°n siendo atendidas de manera eficiente, lo cual puede afectar negativamente la experiencia del cliente.

Distribuci√≥n de Llamadas Entrantes por Operador: En este gr√°fico, podemos ver la distribuci√≥n del volumen de llamadas entrantes entre los operadores. Si algunos operadores manejan muchas m√°s llamadas entrantes que otros, podr√≠a haber un desequilibrio en la asignaci√≥n de las llamadas. Esto podr√≠a indicar que algunos operadores est√°n sobrecargados, mientras que otros est√°n subutilizados. Un ajuste en la distribuci√≥n de las llamadas entrantes podr√≠a mejorar la eficiencia general del equipo.

Distribuci√≥n de Llamadas Salientes por Operador: La mayor√≠a de los puntos est√°n cerca de la parte inferior del eje Y, indicando un n√∫mero bajo de llamadas salientes para la mayor√≠a de los operadores. Sin embargo, hay algunos operadores que realizan un n√∫mero significativamente alto de llamadas salientes, lo que sugiere una variabilidad en la carga de trabajo asignada.

### Identificar y clasificar a los operadores ineficaces seg√∫n las m√©tricas definidas.



```python
# Definir umbrales para considerar a un operador como ineficaz
umbral_llamadas_perdidas = 50
umbral_tiempo_espera_promedio = 5  # en minutos
umbral_llamadas_entrantes = 10
umbral_llamadas_salientes = 10


def es_ineficaz(row):
    if (row['llamadas_perdidas'] > umbral_llamadas_perdidas or
        row['tiempo_espera_promedio'] > umbral_tiempo_espera_promedio or
        row['llamadas_entrantes'] < umbral_llamadas_entrantes or
        row['llamadas_salientes'] < umbral_llamadas_salientes):
        return True
    else:
        return False

metricas['ineficaz'] = metricas.apply(es_ineficaz, axis=1)

operadores_ineficaces = metricas[metricas['ineficaz'] == True]

print("Operadores Ineficaces:")
print(operadores_ineficaces)

```

### Determinar las pruebas estad√≠sticas apropiadas para cada hip√≥tesis.

Hip√≥tesis Nula: No existe una relaci√≥n significativa entre la cantidad de llamadas entrantes perdidas y la eficacia del operador.

Hip√≥tesis Alternativa: Existe una relaci√≥n significativa entre la cantidad de llamadas entrantes perdidas y la eficacia del operador. 

```python
# Crear grupos
eficaces = metricas[metricas['ineficaz'] == False]
ineficaces = metricas[metricas['ineficaz'] == True]

alpha = 0.05

# Comparar las llamadas perdidas entre operadores eficaces e ineficaces
result = t_stat, p_value = ttest_ind(eficaces['llamadas_perdidas'], ineficaces['llamadas_perdidas'])

print(f'Prueba t de dos muestras independientes para llamadas perdidas:')
print(f'p-value: {p_value}')

if result.pvalue < alpha:
    print("\nRechazamos la hip√≥tesis nula")
else:
    print("\nNo podemos rechazar la hip√≥tesis nula")
```

```python
metricas_a_comparar = ['llamadas_perdidas', 'tiempo_espera_promedio', 'llamadas_entrantes', 'llamadas_salientes']

for metrica in metricas_a_comparar:
    t_stat, p_value = ttest_ind(eficaces[metrica], ineficaces[metrica])
    print(f'\nPrueba t de dos muestras independientes para {metrica}:')
    print(f't-statistic: {t_stat}')
    print(f'p-value: {p_value}')
```

Dado que el p-value es mucho mayor que 0.05, no podemos rechazar la hip√≥tesis nula. Esto significa que, seg√∫n los datos proporcionados, no hay una diferencia significativa en las llamadas perdidas entre los operadores eficaces e ineficaces. En otras palabras, los resultados sugieren que la cantidad de llamadas perdidas no est√° relacionada significativamente con la eficacia del operador en este contexto.
